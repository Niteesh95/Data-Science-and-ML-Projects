### Simple Moving Average
let us consider a window size = 10, the SMA  is given by taking average of the first 10 records and putting the result in the 11th record. next for the 12th record, the window is shifted on space below, i.e the average of 2 -11th records is taken.

```python
plt.rc("figure", figsize=(15,6))
plt.figure(figsize=(15,6))
# plt.subplot(212)
plot_pacf(df_6months['Close'])
plt.show()
```

We perfrom the Dickey Fuller test to support our null hypothesis, that the dataset in not stationary. We take into account the last 6months and 1 year data.

```python

from statsmodels.tsa.stattools import adfuller
ADF_result = adfuller(df_6months['Close'])   
print(f'ADF Statistic over past 6 months: {ADF_result[0]}') 
print(f'p-value over past 6 months: {ADF_result[1]}')
```

```python

from statsmodels.tsa.stattools import adfuller
ADF_result = adfuller(df_1year['Close'])   
print(f'ADF Statistic over past 1 year: {ADF_result[0]}') 
print(f'p-value over past 1 year: {ADF_result[1]}')
```

As we can see the ADF statistic of -1.54 and a p-value of 0.51 for 6 months data and -1.91 and 0.32 for 1 year data. Therefore, our time series is not stationary and we must apply transformations to make it stationary. To do so, we will apply the first-order differencing, n=1

```python
# df_6months_diff= np.diff(df_6months['Close'], n=1)
df_6months['close_1diff'] = df_6months['Close'] - df_6months['Close'].shift(1)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.rc("figure", figsize=(15,6))
plt.figure(figsize=(15,6))
plot_acf(df_6months['close_1diff'].dropna())
plt.show()
```

```python
df_6months['close_1diff'].plot(figsize=(10,5))
```

As we can see after differencing by 1 order, the data seems to be somewhat stationary.

## LSTM 

```python
df = DataReader('AAPL', data_source='yahoo', start='2012-11-01', end=datetime.now())
df
```

```python
plt.figure(figsize=(15,5))
plt.title('Apple Inc. Close Price History', fontsize=18)
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=15)
plt.ylabel('Close Price USD ($)', fontsize=15)
plt.grid()
plt.show()
```

```python
data = df.filter(['Close'])
dataset = data.values
train_len = int(np.ceil( len(dataset) * .90 ))

train_len
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range=(0,1))
scaled_data = sc.fit_transform(dataset)

scaled_data
```

```python
train_data = scaled_data[0:int(train_len), :]
X_train = []
y_train = []

for i in range(60, len(train_data)):
    X_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(X_train)
        print(y_train)
        print()
        
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshape the data
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
# !pip install tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(256, return_sequences=True, input_shape= (X_train.shape[1], 1)))
model.add(LSTM(128, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=1, epochs=1)
```

```python
test_data = scaled_data[train_len - 60: , :]
# Create the data sets x_test and y_test
X_test = []
y_test = dataset[train_len:, :]
for i in range(60, len(test_data)):
    X_test.append(test_data[i-60:i, 0])
    
X_test = np.array(X_test)

# Reshape the data
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1 ))
```

```python
predictions = model.predict(X_test)
predictions = sc.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
8.345904016049182
# Plot the data
train = data[:train_len]
valid = data[train_len:]
valid['Predictions'] = predictions
# Visualize the data
plt.figure(figsize=(16,6))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
