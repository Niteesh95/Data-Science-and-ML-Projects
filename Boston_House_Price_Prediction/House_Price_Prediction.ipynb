### Data Imputation and Cleaning

```python
categorical_features =[feature for feature in df_train.columns if df_train[feature].dtypes == 'O']

continuous_df = ['LotFrontage',
 'LotArea',
 'MasVnrArea',
 'BsmtFinSF1',
 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 '1stFlrSF',
 '2ndFlrSF',
 'GrLivArea',
 'GarageArea',
 'WoodDeckSF',
 'OpenPorchSF',
 'EnclosedPorch',
 'ScreenPorch',
 'SalePrice']

discrete_feat = ['MSSubClass',
 'OverallQual',
 'OverallCond',
 'LowQualFinSF',
 'BsmtFullBath',
 'BsmtHalfBath',
 'FullBath',
 'HalfBath',
 'BedroomAbvGr',
 'KitchenAbvGr',
 'TotRmsAbvGrd',
 'Fireplaces',
 'GarageCars',
 '3SsnPorch',
 'PoolArea',
 'MiscVal',
 'MoSold',
 'YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
```

Replace all the null values in the continuous numerical features with the median value. Since these variables are skewed and nor normal, we will replace with median and not mean.

```python
df_train.info()
```

Replace all the null values in the features other than continuous numerical feature, with the mode (maximum frequency) value.

```python
for feature in discrete_feat:
#     print(feature)
    if feature not in 'SalePrice':
        df_train[feature] = df_train[feature].fillna(df_train[feature].mode()[0])
        df_test[feature] = df_test[feature].fillna(df_test[feature].mode()[0])
```

```python
for feature in categorical_features:
#     print(feature)
    if feature not in 'SalePrice':
        df_train[feature] = df_train[feature].fillna(df_train[feature].mode()[0])
        df_test[feature] = df_test[feature].fillna(df_test[feature].mode()[0])
```

```python
continous_numerical_features = [feature for feature in continuous_df if feature not in 'SalePrice']
```

```python
for feature in continous_numerical_features:

    df_train[feature] = df_train[feature].fillna(df_train[feature].median())
    df_test[feature] = df_test[feature].fillna(df_test[feature].median())
```

```python
#Outlier detection
for col in continuous_df:
    q1 = df_train[col].quantile(0.25)
    q3 = df_train[col].quantile(0.75)
    IQR = q3-q1
    min_Value = q1-1.5*IQR
    max_Value = q3+1.5*IQR

    outliers = []
    for i in df_train[col]:
        #print(i)
        if i< min_Value or i>max_Value:
            outliers.append(i)
    print(f"Outliers in {col}:",len(outliers))
```

There are a few outliers in the data, We replace the outliers with the upper and lower percentile values, i.e. top 5% of the data replaced by the 95th percentile value and same for lower 5% with value at 5th percentile. This is the most recommended method to deal with outliers when we have less data.

```python
from scipy.stats.mstats import winsorize
for col_name in continuous_df:    
    df_train[col_name] = winsorize(df_train[col_name], limits=[0.05, 0.05],inclusive=(True, True))
```
